```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulação de dados socioeconômicos da comunidade atendida pela ONG
data = {
    'Bairro': ['Vila Nova', 'Jardim das Flores', 'Parque Verde', 'Centro', 'Subúrbio'],
    'Renda Média (R$)': [1200, 2500, 1800, 3500, 900],
    'Nível de Escolaridade (%)': [25, 40, 35, 60, 15],  # Percentual de pessoas com ensino médio completo
    'População': [800, 500, 600, 1000, 400],
    'Apoio Social (%)': [60, 50, 55, 70, 80]  # Percentual de famílias que recebem apoio da ONG
}

# Criando um DataFrame com os dados
df = pd.DataFrame(data)

# ----- Gráfico de Barras: Comparação de Renda Média e Apoio Social por Bairro -----
plt.figure(figsize=(10, 6))

# Gráfico de barras da renda média
plt.bar(df['Bairro'], df['Renda Média (R$)'], color='skyblue', label='Renda Média (R$)')

# Adicionando título e legendas
plt.title('Renda Média por Bairro da Comunidade', fontsize=14)
plt.xlabel('Bairro', fontsize=12)
plt.ylabel('Renda Média (R$)', fontsize=12)
plt.xticks(rotation=45)
plt.legend()

# Exibindo o gráfico
plt.tight_layout()
plt.show()

# ----- Gráfico de Dispersão: Relação entre Nível de Escolaridade e Renda Média -----
plt.figure(figsize=(10, 6))

# Gráfico de dispersão (scatter plot)
sns.scatterplot(x=df['Nível de Escolaridade (%)'], y=df['Renda Média (R$)'], size=df['População'],
                sizes=(50, 500), hue=df['Bairro'], palette='viridis', legend=True)

# Adicionando título e legendas
plt.title('Relação entre Nível de Escolaridade e Renda Média', fontsize=14)
plt.xlabel('Nível de Escolaridade (%)', fontsize=12)
plt.ylabel('Renda Média (R$)', fontsize=12)

# Exibindo o gráfico
plt.tight_layout()
plt.show()

# ----- Gráfico de Barras: Percentual de Apoio Social por Bairro -----
plt.figure(figsize=(10, 6))

# Gráfico de barras do apoio social
sns.barplot(x=df['Bairro'], y=df['Apoio Social (%)'], palette='coolwarm')

# Adicionando título e legendas
plt.title('Percentual de Apoio Social por Bairro', fontsize=14)
plt.xlabel('Bairro', fontsize=12)
plt.ylabel('Apoio Social (%)', fontsize=12)
plt.xticks(rotation=45)

# Exibindo o gráfico
plt.tight_layout()
plt.show()
```
