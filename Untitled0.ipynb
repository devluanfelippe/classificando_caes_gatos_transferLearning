```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Simulao de dados socioeconmicos da comunidade atendida pela ONG
data = {
    'Bairro': ['Vila Nova', 'Jardim das Flores', 'Parque Verde', 'Centro', 'Subrbio'],
    'Renda Mdia (R$)': [1200, 2500, 1800, 3500, 900],
    'Nvel de Escolaridade (%)': [25, 40, 35, 60, 15],  # Percentual de pessoas com ensino mdio completo
    'Populao': [800, 500, 600, 1000, 400],
    'Apoio Social (%)': [60, 50, 55, 70, 80]  # Percentual de famlias que recebem apoio da ONG
}

# Criando um DataFrame com os dados
df = pd.DataFrame(data)

# ----- Grfico de Barras: Comparao de Renda Mdia e Apoio Social por Bairro -----
plt.figure(figsize=(10, 6))

# Grfico de barras da renda mdia
plt.bar(df['Bairro'], df['Renda Mdia (R$)'], color='skyblue', label='Renda Mdia (R$)')

# Adicionando ttulo e legendas
plt.title('Renda Mdia por Bairro da Comunidade', fontsize=14)
plt.xlabel('Bairro', fontsize=12)
plt.ylabel('Renda Mdia (R$)', fontsize=12)
plt.xticks(rotation=45)
plt.legend()

# Exibindo o grfico
plt.tight_layout()
plt.show()

# ----- Grfico de Disperso: Relao entre Nvel de Escolaridade e Renda Mdia -----
plt.figure(figsize=(10, 6))

# Grfico de disperso (scatter plot)
sns.scatterplot(x=df['Nvel de Escolaridade (%)'], y=df['Renda Mdia (R$)'], size=df['Populao'],
                sizes=(50, 500), hue=df['Bairro'], palette='viridis', legend=True)

# Adicionando ttulo e legendas
plt.title('Relao entre Nvel de Escolaridade e Renda Mdia', fontsize=14)
plt.xlabel('Nvel de Escolaridade (%)', fontsize=12)
plt.ylabel('Renda Mdia (R$)', fontsize=12)

# Exibindo o grfico
plt.tight_layout()
plt.show()

# ----- Grfico de Barras: Percentual de Apoio Social por Bairro -----
plt.figure(figsize=(10, 6))

# Grfico de barras do apoio social
sns.barplot(x=df['Bairro'], y=df['Apoio Social (%)'], palette='coolwarm')

# Adicionando ttulo e legendas
plt.title('Percentual de Apoio Social por Bairro', fontsize=14)
plt.xlabel('Bairro', fontsize=12)
plt.ylabel('Apoio Social (%)', fontsize=12)
plt.xticks(rotation=45)

# Exibindo o grfico
plt.tight_layout()
plt.show()
```
